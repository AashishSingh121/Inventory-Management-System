```python
def print_Sales():
    f = open('D:\Aashish Folder\Docs\Sales.json',)
    k=f.read()
    dic = json.loads(k)
    val=dic.keys()
    print("Product Name "," Quantity","Cost","\t  Day-Time")
    for i in val:
        print(dic[i]['Product_name'],"\t\t",dic[i]["Quantity"],"\t",dic[i]["Cost"],"\t",dic[i]["Day & Time"])
    f.close()

def print_product(dic):
    val=dic.keys()
    print("Product Name "," Quantity","Price")
    for i in val:
        print(dic[i]['Product_name'],"\t\t",dic[i]["Quantity"],"\t",dic[i]["Price"])
def updatestock(dic):    
    res="y"
    while(res=="Y" or res=="y"):
        print("What do you want to change?")
        print("1. Add/Remove stock","\n2. Change Price","\n3. Exit")
        u=int(input("Enter choice: "))
        if u==1:
            ide=input("Enter product id:")
            q=int(input("Enter quantity to add/remove:"))
            if int(q)>=0:
                dic[ide]['Quantity']=str(int(dic[ide]['Quantity'])+int(q))
            if int(q)<0:
                if int(dic[ide]['Quantity'])<abs(int(q)):
                    dic[ide]['Quantity']=str(0)
                else:
                    dic[ide]['Quantity']=str(int(dic[ide]['Quantity'])+int(q))
            Updaterecord(dic)
        if u==2:
            ide=input("Enter product id:")
            price=input("Enter New Price:")
            dic[ide]['Price']=price
            Updaterecord(dic)
        if u==3:
            break
        res =input("Do you want to continue(Y/N)?")
    
        
def purchase(dic):
    import datetime
    val=dic.keys()
    print("\nProduct List")
    print("Product ID    Product Name ")
    for i in val:
        print("  ",i,"\t",dic[i]['Product_name'])
    ide=input("\nEnter product id:")
    q=input("Enter Quantity:")
    if int(q)<=int(dic[ide]['Quantity']):
        dic[ide]['Quantity']=str(int(dic[ide]['Quantity'])-int(q))
        print(dic[ide]['Product_name'],"Cost:",int(dic[i]["Price"])*int(q))
        Updaterecord(dic)
        if(int(q)>0):
            time=str(datetime.datetime.now())
            time=time[:19]
            Updatesales(dic[ide]['Product_name'],str(int(dic[i]["Price"])*int(q)),q,time)
        
    else:
        print("Insufficient Quantity")

def Updaterecord(dic):
    f = open('D:\Aashish Folder\Docs\Records.json','w')
    j_object=json.dumps(dic)
    f.write(j_object)
    f.close()

def Updatesales(name,cost,q,time):
    f = open('D:\Aashish Folder\Docs\Sales.json','r')
    k=f.read()
    y = json.loads(k)
    f.close()
    val=list(y.keys())
    n=len(val)
    ide=str(int(val[n-1])+1)
    y[ide]={"Product_name": name, "Quantity": q, "Cost": cost,"Day & Time":time}
    f = open('D:\Aashish Folder\Docs\Sales.json','w')
    j_object=json.dumps(y)
    f.write(j_object)
    f.close()
    
def Addrecord(dic):
    val=list(dic.keys())
    n=len(val)
    ide=str(int(val[n-1])+1)
    name=input("Enter Product Name:")
    q=input("Enter Quantity:")
    c=input("Enter Price:")
    dic[ide]={"Product_name": name, "Quantity": q, "Price": c}
    f = open('D:\Aashish Folder\Docs\Records.json','w')
    j_object=json.dumps(dic)
    f.write(j_object)
    
from IPython.display import clear_output
import json
f = open('D:\Aashish Folder\Docs\Records.json',)
k=f.read()
y = json.loads(k)
res="y"
print("Welcome !!!")
print("\n Menu")
while(res=="y" or res=="Y"):
    
    print(" 1. Make purchase\n","2. View Inventory\n","3. Add Product\n","4. View Sales\n","5. Update Product details\n","6. Exit")
    key=int(input("Enter Choice:"))
    

    if key==1:
        purchase(y)
    elif key==2:
        print_product(y)
    elif key==3:
        Addrecord(y)
    elif key==4:
        print_Sales()
    elif key==5:
        updatestock(y)
        continue
    elif key==6:
        break
    else:
        print("Invalid Input")
    res =input("Do you want to continue(Y/N)?")
    clear_output()
    
    
f.close()    
```
